```python
x=df.iloc[:,:-1]
x
```

```python
y=df["Personality"]
y
```

```python
from tensorflow.keras.models import Sequential
from keras.layers import Dense,Dropout,BatchNormalization
from keras.callbacks import EarlyStopping
```

```python
clf=Sequential()
```

```python
clf.add(Dense(units=15,kernel_initializer='he_normal',activation='relu'))
clf.add(Dense(units=10,kernel_initializer='he_normal',activation='relu'))
clf.add(Dense(units=1,kernel_initializer='glorot_uniform',activation='sigmoid'))
clf.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```
